Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

🎉**BONUS**: can you adjust the `x-axis` labels to have "comma" formatting? (**Hint:** use the `comma_format()` function from the `scales` package.)

```r
options(repr.plot.res = 300, repr.plot.height = 4, repr.plot.width = 4.5)
p = ggplot(penguins, aes(x = body_mass_g))
p = p + geom_histogram(bins = 40, fill = 'cadetblue', color = "cadetblue4")
p = p + facet_wrap( ~ year, nrow = 3) + 
        theme_minimal() +
        labs(x = "Body mass in grams", y = "Number of Penguins" ) +
        theme(strip.background = element_rect(color="transparent", fill="transparent"), 
             strip.text.x = element_text(size = 9, color = "grey50", face = "bold", hjust = 0.95), 
             axis.title.x = element_text(size = 10), axis.title.y = element_text(size = 10))
p = p + labs(title = "Histograms of penguin body mass for measurements taken from 2007-2009") +
        theme(plot.title = element_text(size = 8, hjust = 0), plot.title.position = 'plot') +
        scale_x_continuous(labels = scales::comma)
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
penguins |> head()
```

```r
p2 = ggplot(penguins, aes(x = body_mass_g, y = sex))
p2 = p2 + geom_point(position = position_jitter(h = 0.2, w = 0), size = 0.75, alpha = 0.5, color = "bisque4" ) +
     facet_wrap( ~ species, nrow = 3) +
     theme_minimal() +
     labs(x = "Body mass in grams" ) +
     theme(strip.background = element_rect(color="transparent", fill="transparent"), 
             strip.text.x = element_text(size = 9, color = "grey50", face = "bold", hjust = 0.15), 
             axis.title.x = element_text(size = 10), 
            axis.title.y = element_blank()) +
     scale_x_continuous(labels = scales::comma)
p2
```
